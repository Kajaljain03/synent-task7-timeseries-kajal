```python
# Trend Analysis Chart
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(df['Date'], df['Close'])

plt.title("Apple Stock Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Closing Price")

plt.show()
```

```python
# Moving Average Analysis
# Create 30-day moving average
df['Moving_Avg_30'] = df['Close'].rolling(window=30).mean()

# Plot
plt.figure(figsize=(12,6))

plt.plot(df['Date'], df['Close'], label='Closing Price')
plt.plot(df['Date'], df['Moving_Avg_30'], label='30-Day Moving Average')

plt.title("Closing Price vs 30-Day Moving Average")

plt.xlabel("Date")
plt.ylabel("Price")

plt.legend()

plt.show()
```

```python
# Highest & Lowest Closing Price
highest_price = df['Close'].max()
lowest_price = df['Close'].min()

print("Highest Closing Price:", highest_price)
print("Lowest Closing Price:", lowest_price)
```

```python
# Correlation Analysis
correlation = df[['Open', 'High', 'Low', 'Close', 'Volume']].corr()

correlation
```

```python
# Average Monthly trend
df['Month'] = df['Date'].dt.month

monthly_avg = df.groupby('Month')['Close'].mean()

monthly_avg.plot(kind='bar', figsize=(10,5))

plt.title("Average Monthly Closing Price")

plt.xlabel("Month")
plt.ylabel("Average Closing Price")

plt.show()
```
